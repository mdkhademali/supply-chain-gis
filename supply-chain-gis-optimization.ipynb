# Section 1: Introduction

## 1.1 Supply Chain Optimization Conceptual Foundation

**Supply Chain Optimization (SCO)** is the systematic application of mathematical, computational, and analytical methods to improve the efficiency, resilience, and cost-effectiveness of logistics networks. A supply chain spans all activities from raw-material procurement through production, warehousing, distribution, and final delivery.

Modern supply chains are characterised by:
- **Global geographic extent** spanning multiple continents and transport modes
- **Multi-modal transportation** road, rail, maritime, and air
- **Dynamic demand** shaped by seasonality, economics, and exogenous events
- **Risk exposure** to pandemics, climate events, and geopolitical disruptions

Mathematically, total cost minimisation is written:

$$Z = \sum_{i}\sum_{j} c_{ij} x_{ij} + \sum_{j} f_j y_j + \sum_{j} h_j I_j$$

where $c_{ij}$ = unit transport cost, $x_{ij}$ = flow, $f_j$ = fixed facility cost, $y_j \in \{0,1\}$ = open/close decision, $h_j$ = holding cost, $I_j$ = inventory level.

---

## 1.2 GIS in Logistics

**Geographic Information Systems (GIS)** provide the spatial backbone for supply-chain decision-making. Unlike abstract network models, GIS-powered logistics:

1. Represents **real-world geography**: actual road networks, terrain, land use
2. Enables **spatial queries**: proximity, overlay, containment, buffering
3. Supports **multi-layer analysis**: combining population, infrastructure, and demand
4. Facilitates **communication**: through universally understood maps

---

## 1.3 Spatial Intelligence and Network Science

**Spatial Intelligence** integrates geospatial data, AI, and domain knowledge:

| Centrality Measure | Formula | Logistics Use |
|-------------------|---------|--------------|
| Degree | $C_D(v)=\frac{\deg(v)}{n-1}$ | Hub identification |
| Betweenness | $C_B(v)=\sum_{s\neq v\neq t}\frac{\sigma_{st}(v)}{\sigma_{st}}$ | Bottleneck detection |
| Closeness | $C_C(v)=\frac{n-1}{\sum_u d(v,u)}$ | Accessibility measure |
| PageRank | $PR(v)=\frac{1-d}{n}+d\sum_u\frac{PR(u)}{\deg(u)}$ | Network importance |

---

## 1.4 Transportation Geography

Key spatial-interaction theories:
- **Gravity Model**: $T_{ij}=k\frac{P_i P_j}{d_{ij}^\alpha}$, trip generation between zones
- **Central Place Theory**: hierarchical distribution of facilities
- **Accessibility Index**: (Hansen): $A_i=\sum_j O_j e^{-\beta d_{ij}}$

# Section 2: Literature Review

## 2.1 GIS in Logistics and Supply Chain Management

The intersection of GIS and supply-chain management has been a rich area of inquiry since the 1990s. **Ghosh & McLafferty (1987)** established foundational principles for spatial retail analysis; **Daskin (1995)** formalised network facility location problems underpinning modern GIS logistics.

### 2.1.1 Facility Location

The Facility Location Problem (FLP) was formalised by **ReVelle & Swain (1970)** as the P-Median model. **Church & ReVelle (1974)** introduced the Maximum Covering Location Problem (MCLP). Contemporary extensions incorporate stochastic demand, multi-objective optimisation, and dynamic relocation.

### 2.1.2 Supply Chain Resilience

**Christopher & Peck (2004)** identified four supply-chain risk categories: supply, operational, demand, and environmental. Spatial analysis maps vulnerability hotspots (flood zones, conflict areas) and redundancy paths.

### 2.1.3 AI in Transportation

**Vlahogianni et al. (2015)** reviewed deep learning for traffic prediction. **Graph Convolutional Networks** applied to road networks (Yao et al., 2018) represent the GeoAI frontier.

## 2.3 Research Gaps and Motivation

1. **Integration gap**: few studies combine FLP + VRP + ML + Resilience in one framework

````python
# Section 4.3 | Synthetic Demand Data Generation
np.random.seed(SEED)

def clustered_pts(n, k=6, spread=0.022):
    centers_lat=np.random.uniform(LAT_MIN+.04,LAT_MAX-.04,k)
    centers_lon=np.random.uniform(LON_MIN+.04,LON_MAX-.04,k)
    w=np.random.dirichlet(np.ones(k))
    pts=[]
    for _ in range(n):
        c=np.random.choice(k,p=w)
        pts.append((np.clip(np.random.normal(centers_lat[c],spread),LAT_MIN,LAT_MAX),
                    np.clip(np.random.normal(centers_lon[c],spread),LON_MIN,LON_MAX)))
    return np.array(pts)

# 1. Customers (n=300)
N_C=300
cc=clustered_pts(N_C,k=7)
customers=gpd.GeoDataFrame({
    'customer_id':range(N_C),
    'latitude':cc[:,0],'longitude':cc[:,1],
    'demand_kg':np.random.lognormal(3.5,.8,N_C).round(1),
    'demand_units':np.random.randint(1,50,N_C),
    'priority':np.random.choice(['High','Medium','Low'],N_C,p=[.2,.5,.3]),
    'customer_type':np.random.choice(
        ['Retail','Restaurant','Office','Residential','Healthcare'],
        N_C,p=[.3,.2,.25,.15,.1]),
    'monthly_revenue_usd':np.random.lognormal(7,1.2,N_C).round(0),
},geometry=[Point(lon,lat) for lat,lon in cc],crs='EPSG:4326')

# 2. Warehouse Candidates (n=20)
N_W=20
wc=clustered_pts(N_W,k=4,spread=0.04)
warehouses=gpd.GeoDataFrame({
    'warehouse_id':range(N_W),
    'name':[f'WH-{i+1:02d}' for i in range(N_W)],
    'latitude':wc[:,0],'longitude':wc[:,1],
    'capacity_tons':np.random.choice([50,100,200,500],N_W,p=[.3,.4,.2,.1]),
    'fixed_cost_annual_usd':np.random.randint(50000,500000,N_W),
    'operating_cost_per_day':np.random.randint(500,5000,N_W),
    'has_cold_storage':np.random.choice([True,False],N_W,p=[.3,.7]),
    'road_access_score':np.random.uniform(.5,1.,N_W).round(2),
},geometry=[Point(lon,lat) for lat,lon in wc],crs='EPSG:4326')

# 3. Time Series (2 years daily)
n_days=730
dates=pd.date_range('2022-01-01',periods=n_days,freq='D')
t=np.arange(n_days)
trend=np.linspace(1000,1400,n_days)
weekly=150*np.sin(2*np.pi*t/7)
annual=200*np.sin(2*np.pi*t/365-np.pi/2)
spikes=np.zeros(n_days); spikes[330]=600; spikes[359]=750
noise=np.random.normal(0,80,n_days)
demand_series=pd.DataFrame({
    'date':dates,
    'demand_kg':np.maximum(0,trend+weekly+annual+spikes+noise).round(1),
    'day_of_week':dates.dayofweek,'month':dates.month,
    'quarter':dates.quarter,'is_weekend':(dates.dayofweek>=5).astype(int),
    'day_of_year':dates.dayofyear,'year':dates.year,
})

# 4. Fleet
fleet=pd.DataFrame({
    'vehicle_id':[f'V-{i+1:03d}' for i in range(15)],
    'type':np.random.choice(['Motorbike','Van','Truck','Ref.Truck'],15,p=[.3,.3,.3,.1]),
    'capacity_kg':np.random.choice([50,500,2000,5000],15,p=[.3,.3,.3,.1]),
    'cost_per_km':np.random.choice([.05,.15,.35,.55],15,p=[.3,.3,.3,.1]),
    'fuel_type':np.random.choice(['Petrol','Diesel','Electric'],15,p=[.3,.5,.2]),
})

print(f" Customers: {len(customers):,} | Warehouses: {len(warehouses)}")
print(f"   Total demand: {customers['demand_kg'].sum():,.0f} kg")
print(f"   Time series: {len(demand_series):,} days | Fleet: {len(fleet)} vehicles")
print("\nFleet summary:")
print(fleet.groupby('type')[['capacity_kg','cost_per_km']].mean().round(2))
````

# Section 5: Exploratory Spatial Data Analysis (ESDA)
## 5.1 Theory

ESDA examines the spatial distribution, autocorrelation, and pattern of geographic data before formal modelling.

**Kernel Density Estimation:**
$$\hat{f}(x)=\frac{1}{nh}\sum_{i=1}^{n}K\!\left(\frac{x-x_i}{h}\right)$$

**Moran's I (spatial autocorrelation):**
$$I=\frac{n}{\sum_i\sum_j w_{ij}}\cdot\frac{\sum_i\sum_j w_{ij}(y_i-\bar y)(y_j-\bar y)}{\sum_i(y_i-\bar y)^2}$$

$I>0$ indicates positive spatial autocorrelation (similar values cluster together). Tobler's First Law states: *"near things are more related than distant things."*

# Section 6: Road Network Analysis
## 6.1 Graph Theory Foundations

Transportation networks are directed weighted graphs $G=(V,E,W)$ where nodes are intersections and edges are road segments.

### Shortest Path Algorithms

**Dijkstra's Algorithm**, $O((V+E)\log V)$:
```
dist[source]=0; dist[others]=∞
While queue not empty:
    (cost,u)=dequeue_min()
    For each neighbor v:
        if cost+w(u,v) < dist[v]: update & enqueue
```

**A\*** improves Dijkstra using a geodesic heuristic $h(v)$, reducing search space.

### Accessibility Analysis

**Hansen-type cumulative accessibility:**
$$A_i = \sum_j O_j \cdot f(t_{ij})$$

where $O_j$ is opportunity at $j$ and $f(t_{ij})$ is an impedance decay function.


````python
# Section 6 | Network Analysis
# Prepare analysis graph
G_und = G.to_undirected() if G.is_directed() else G
max_nodes=500
if G_und.number_of_nodes()>max_nodes:
    G_s = G_und.subgraph(list(G_und.nodes())[:max_nodes]).copy()
else:
    G_s = G_und.copy()

# Node positions
pos={}
for n in G_s.nodes():
    pos[n]=(G_s.nodes[n].get('x',0),G_s.nodes[n].get('y',0))

print(f"Network for analysis: {G_s.number_of_nodes()} nodes, {G_s.number_of_edges()} edges")

# Centrality measures
print("Computing centrality measures …")
deg_c  = nx.degree_centrality(G_s)
bet_c  = nx.betweenness_centrality(G_s, k=min(80,len(G_s)))
clo_c  = nx.closeness_centrality(G_s)
pr     = nx.pagerank(G_s, alpha=.85, max_iter=100)
print(" Centrality computed")

cdf=pd.DataFrame({
    'node':list(G_s.nodes()),
    'degree':    [deg_c[n] for n in G_s.nodes()],
    'betweenness':[bet_c[n] for n in G_s.nodes()],
    'closeness': [clo_c[n] for n in G_s.nodes()],
    'pagerank':  [pr[n]    for n in G_s.nodes()],
})
for col in ['degree','betweenness','closeness','pagerank']:
    cdf[col+'_norm']=MinMaxScaler().fit_transform(cdf[[col]]).ravel()

cdf['composite']=(cdf['degree_norm']*.25+cdf['betweenness_norm']*.35+
                  cdf['closeness_norm']*.25+cdf['pagerank_norm']*.15)

# Shortest path
nl=list(G_s.nodes())
src,tgt=nl[0],nl[len(nl)//2]
try:
    sp=nx.shortest_path(G_s,source=src,target=tgt)
    sp_len=nx.shortest_path_length(G_s,source=src,target=tgt)
    print(f"\n  Shortest path: {len(sp)} nodes, {sp_len} hops")
except nx.NetworkXNoPath:
    sp=[]; print("\n  No path between selected nodes")

print("\nCentrality statistics:")
print(cdf[['degree','betweenness','closeness','pagerank']].describe().round(4))
````

# Section 7: Warehouse Location Optimization
## 7.1 Facility Location Problem Theory

### P-Median Model

$$\min\sum_{i\in I}\sum_{j\in J}d_{ij}w_i x_{ij}$$
s.t. $\sum_j x_{ij}=1\;\forall i$, $x_{ij}\le y_j$, $\sum_j y_j=p$, $x_{ij},y_j\in\{0,1\}$

### Set Covering Location Problem

$$\min\sum_{j\in J}y_j$$
s.t. $\sum_{j\in N_i}y_j\ge 1\;\forall i$, where $N_i=\{j:d_{ij}\le R\}$

The P-Median minimises **total weighted distance**; Set Covering minimises **number of facilities** while ensuring every customer is covered within threshold $R$.


````python
# Section 7 | Distance Matrix
def haversine(lats1,lons1,lats2,lons2):
    R=6371.
    la1=np.radians(lats1)[:,None]; lo1=np.radians(lons1)[:,None]
    la2=np.radians(lats2)[None,:]; lo2=np.radians(lons2)[None,:]
    a=np.sin((la2-la1)/2)**2+np.cos(la1)*np.cos(la2)*np.sin((lo2-lo1)/2)**2
    return 2*R*np.arcsin(np.sqrt(a))

DM=haversine(customers['latitude'].values,customers['longitude'].values,
             warehouses['latitude'].values,warehouses['longitude'].values)
demands=customers['demand_kg'].values
print(f"Distance matrix: {DM.shape}  (customers × warehouses)")
print(f"Min: {DM.min():.2f} km | Max: {DM.max():.2f} km | Mean: {DM.mean():.2f} km")
````

````python
# Section 7 | P-Median Greedy Heuristic
def p_median(DM,demands,p):
    n_c,n_f=DM.shape; selected=[]; remaining=list(range(n_f))
    for step in range(p):
        best_j,best_obj=None,np.inf
        for j in remaining:
            trial=selected+[j]
            obj=(DM[:,trial].min(axis=1)*demands).sum()
            if obj<best_obj: best_obj,best_j=obj,j
        selected.append(best_j); remaining.remove(best_j)
        print(f"  Opened {warehouses.iloc[best_j]['name']}  obj={best_obj:,.1f}")
    asgn_local=DM[:,selected].argmin(axis=1)
    asgn_global=np.array(selected)[asgn_local]
    return selected,asgn_global,(DM[:,selected].min(axis=1)*demands).sum()

results={}
for p in [3,4,5]:
    print(f"\n🔍 P-Median p={p}:")
    sel,asgn,obj=p_median(DM,demands,p)
    results[p]={'sel':sel,'asgn':asgn,'obj':obj,
                'avg':DM[:,sel].min(axis=1).mean(),
                'max':DM[:,sel].min(axis=1).max()}

print(f"\n{'p':>3}|{'Objective (km·kg)':>19}|{'Avg (km)':>10}|{'Max (km)':>10}")
print("-"*46)
for p,r in results.items():
    print(f"{p:>3}|{r['obj']:>19,.1f}|{r['avg']:>10.2f}|{r['max']:>10.2f}")

p3o,p4o,p5o=results[3]['obj'],results[4]['obj'],results[5]['obj']
print(f"\nMarginal benefit 3→4: {(p3o-p4o)/p3o*100:.1f}% cost reduction")
print(f"Marginal benefit 4→5: {(p4o-p5o)/p4o*100:.1f}% cost reduction")
print("→ p=4 gives best cost–coverage trade-off")

P_OPT=4
selected_wh=results[P_OPT]['sel']
assignments=results[P_OPT]['asgn']
````

````python
# Section 7 | Set Covering + Visualisation
def set_cover(DM,thresh):
    cov=(DM<=thresh).astype(int); covered=np.zeros(DM.shape[0],dtype=bool); sel=[]
    while not covered.all():
        nc=np.array([(cov[:,j]&~covered).sum() for j in range(DM.shape[1])])
        if nc.max()==0: break
        bj=nc.argmax(); sel.append(bj); covered|=cov[:,bj].astype(bool)
    return sel,covered.sum()/DM.shape[0]

print("Set Covering Analysis:")
print(f"{'Threshold (km)':>16}|{'Facilities':>12}|{'Coverage %':>12}")
print("-"*42)
for th in [2,3,5,8,10]:
    s,cr=set_cover(DM,th)
    print(f"{th:>16}|{len(s):>12}|{cr*100:>11.1f}%")

# Visualisation
WC=['#ff6b6b','#ffd93d','#6bcb77','#4d96ff','#c77dff']
fig,axes=plt.subplots(1,2,figsize=(18,8)); fig.patch.set_facecolor('#0d1117')

for ax,pv in zip(axes,[3,4]):
    ax.set_facecolor('#0d1117')
    sel=results[pv]['sel']; asgn=results[pv]['asgn']
    for k,wi in enumerate(sel):
        mask=(asgn==wi)
        ax.scatter(customers.loc[mask,'longitude'],customers.loc[mask,'latitude'],
                   c=WC[k%5],s=28,alpha=.65,label=f'Zone {k+1} ({mask.sum()})')
        zone_c=customers[mask].head(8)
        wh=warehouses.iloc[wi]
        for _,row in zone_c.iterrows():
            ax.plot([row['longitude'],wh['longitude']],[row['latitude'],wh['latitude']],
                    color=WC[k%5],alpha=.2,lw=.7)
    ax.scatter(warehouses['longitude'],warehouses['latitude'],
               marker='s',s=60,c='#30363d',zorder=4,edgecolors='white',lw=.5)
    for k,wi in enumerate(sel):
        wh=warehouses.iloc[wi]
        ax.scatter([wh['longitude']],[wh['latitude']],marker='*',s=450,
                   c=WC[k%5],zorder=5,edgecolors='white',lw=1)
        ax.annotate(wh['name'],(wh['longitude'],wh['latitude']),
                    xytext=(7,7),textcoords='offset points',color='white',
                    fontsize=8,fontweight='bold',
                    bbox=dict(boxstyle='round,pad=.2',fc='#161b22',alpha=.8))
    ax.set_title(f'P-Median p={pv}  |  Obj={results[pv]["obj"]:,.0f} km·kg
'
                 f'Avg dist={results[pv]["avg"]:.2f} km  Max dist={results[pv]["max"]:.2f} km',
                 color='white',fontsize=11,pad=8)
    ax.tick_params(colors='#8b949e',labelsize=8)
    ax.legend(fontsize=7,facecolor='#161b22',labelcolor='white')
    for sp in ax.spines.values(): sp.set_edgecolor('#30363d')

plt.suptitle('Section 7 | Warehouse Location Optimisation (P-Median)',
             fontsize=15,color='white',fontweight='bold',y=1.01)
plt.tight_layout(pad=2.); plt.savefig('warehouse_opt.png',dpi=150,bbox_inches='tight',facecolor='#0d1117')
plt.show(); print("Warehouse optimisation complete")
````

# Section 8: Vehicle Routing Problem (VRP)
## 8.1 CVRP Formulation

**Capacitated VRP:** Given depot, $n$ customers with demands $d_i$, $m$ vehicles with capacity $Q$:

$$\min\sum_k\sum_i\sum_j c_{ij}x_{ijk}$$

subject to each customer visited exactly once and $\sum_i d_i x_{ijk}\le Q\;\forall k$.

### Clarke-Wright Savings Algorithm (1964)

**Savings** of combining routes via $i$ and $j$:
$$s_{ij}=c_{0i}+c_{0j}-c_{ij}$$

Merge routes in descending savings order until capacity constraint binds.  
Time complexity: $O(n^2\log n)$, highly practical for real logistics instances.


````python
# Section 8 | CVRP — Clarke-Wright
depot_idx=selected_wh[0]
depot=warehouses.iloc[depot_idx]
near_mask=DM[:,depot_idx]<=5.0
vrp_c=customers[near_mask].copy().reset_index(drop=True)
V_CAP=500.0
print(f"CVRP Setup")
print(f"   Depot: {depot['name']}  ({depot['latitude']:.4f}, {depot['longitude']:.4f})")
print(f"   Customers in service area: {len(vrp_c)}")
print(f"   Total demand: {vrp_c['demand_kg'].sum():.1f} kg | Vehicle capacity: {V_CAP} kg")
print(f"   Theoretical min vehicles: {int(np.ceil(vrp_c['demand_kg'].sum()/V_CAP))}")

def cw_vrp(dlat,dlon,clats,clons,dems,cap):
    n=len(clats)
    if n==0: return [],np.zeros((1,1))
    alats=np.concatenate([[dlat],clats]); alons=np.concatenate([[dlon],clons])
    dist=haversine(alats,alons,alats,alons)
    savings=sorted([(dist[0,i]+dist[0,j]-dist[i,j],i-1,j-1)
                    for i in range(1,n+1) for j in range(i+1,n+1)],reverse=True)
    routes=[[c] for c in range(n)]
    rdem=[dems[c] for c in range(n)]
    rof={c:c for c in range(n)}
    for s,i,j in savings:
        ri,rj=rof.get(i),rof.get(j)
        if ri is None or rj is None or ri==rj: continue
        ri_r,rj_r=routes[ri],routes[rj]
        if ri_r is None or rj_r is None: continue
        merged=None
        if ri_r[-1]==i and rj_r[0]==j:   merged=ri_r+rj_r
        elif rj_r[-1]==j and ri_r[0]==i: merged=rj_r+ri_r
        elif ri_r[-1]==i and rj_r[-1]==j:merged=ri_r+rj_r[::-1]
        elif ri_r[0]==i and rj_r[0]==j:  merged=ri_r[::-1]+rj_r
        if merged and rdem[ri]+rdem[rj]<=cap:
            nr=len(routes); routes.append(merged)
            routes[ri]=None; routes[rj]=None
            rdem.append(rdem[ri]+rdem[rj])
            for c in merged: rof[c]=nr
    return [r for r in routes if r],dist

if len(vrp_c)>0:
    routes,da=cw_vrp(depot['latitude'],depot['longitude'],
                     vrp_c['latitude'].values,vrp_c['longitude'].values,
                     vrp_c['demand_kg'].values,V_CAP)

    def rdist(route,d):
        if not route: return 0.
        tot=d[0,route[0]+1]
        for k in range(len(route)-1): tot+=d[route[k]+1,route[k+1]+1]
        return tot+d[route[-1]+1,0]

    rm=[]
    for idx,r in enumerate(routes):
        if not r: continue
        dem=vrp_c.iloc[r]['demand_kg'].sum()
        dk=rdist(r,da)
        rm.append({'route_id':idx+1,'n_stops':len(r),'demand_kg':dem,
                   'distance_km':dk,'utilization':dem/V_CAP*100,
                   'cost_usd':dk*.35,'customers':r})
    metrics_df=pd.DataFrame(rm)

    print(f"\n CVRP Solution:")
    print(f"   Routes: {len(metrics_df)} | Total dist: {metrics_df['distance_km'].sum():.2f} km")
    print(f"   Total cost: ${metrics_df['cost_usd'].sum():.2f}")
    print(f"   Avg utilisation: {metrics_df['utilization'].mean():.1f}%")
    print("\nRoute details:")
    print(metrics_df[['route_id','n_stops','demand_kg','distance_km',
                       'utilization','cost_usd']].round(2).to_string(index=False))
else:
    print("⚠️  No customers within 5 km of depot — expanding to 8 km")
    near_mask=DM[:,depot_idx]<=8.0
    vrp_c=customers[near_mask].copy().reset_index(drop=True)
    metrics_df=pd.DataFrame({'route_id':[1],'n_stops':[len(vrp_c)],
                              'demand_kg':[vrp_c['demand_kg'].sum()],
                              'distance_km':[10.],'utilization':[70.],'cost_usd':[3.5],'customers':[[]]})
````

# Section 9: Machine Learning for Demand Forecasting
## 9.1 Theory and Feature Engineering

Demand forecasting enables optimal inventory, fleet sizing, and warehouse allocation.

### Feature Categories

| Category | Features | Rationale |
|----------|---------|-----------|
| **Calendar** | Day-of-week, Month, Quarter | Capture seasonality |
| **Lag** | t-1, t-7, t-14, t-28 | Temporal autocorrelation |
| **Rolling Stats** | 7/14/30-day MA & Std | Trend and volatility |
| **Fourier Terms** | sin/cos(2πk·t/365) | Smooth annual cycles |
| **Event Flags** | Weekend, Month-start/end | Special day effects |

### Model Comparison

| Model | Complexity | Handles Nonlinearity | Interpretable |
|-------|-----------|---------------------|--------------|
| Ridge Regression | Low | ✗ | ✅ |
| Random Forest | Medium | ✅ | Partial |
| XGBoost | High | ✅ | Partial (SHAP) |


````python
# Section 9 | Feature Engineering
df=demand_series.sort_values('date').reset_index(drop=True)

def build_features(df):
    d=df.copy()
    for lag in [1,3,7,14,21,28]: d[f'lag_{lag}']=d['demand_kg'].shift(lag)
    for w in [7,14,30]:
        d[f'rmean_{w}']=d['demand_kg'].shift(1).rolling(w).mean()
        d[f'rstd_{w}']=d['demand_kg'].shift(1).rolling(w).std()
    for k in [1,2,3]:
        d[f'sin_{k}']=np.sin(2*np.pi*k*d['day_of_year']/365)
        d[f'cos_{k}']=np.cos(2*np.pi*k*d['day_of_year']/365)
    d['dow']=d['date'].dt.dayofweek
    d['month']=d['date'].dt.month
    d['quarter']=d['date'].dt.quarter
    d['is_weekend']=(d['date'].dt.dayofweek>=5).astype(int)
    d['mstart']=d['date'].dt.is_month_start.astype(int)
    d['mend']=d['date'].dt.is_month_end.astype(int)
    d['trend']=np.arange(len(d))
    return d.dropna()

dff=build_features(df)
fc=[c for c in dff.columns if c not in ['date','demand_kg','year','day_of_year']]
X=dff[fc].values; y=dff['demand_kg'].values
sp=len(X)-90
Xtr,Xte,ytr,yte=X[:sp],X[sp:],y[:sp],y[sp:]
print(f"Dataset: {len(X):,} samples | Features: {len(fc)}")
print(f"Train: {len(Xtr):,} | Test: {len(Xte):,} (last 90 days)")
````

````python
# Section 9 | Model Training & Evaluation
models_trained={}; preds={}

# 1. Ridge Regression
rr=Pipeline([('sc',StandardScaler()),('m',Ridge(alpha=1.))])
rr.fit(Xtr,ytr); models_trained['Ridge']=rr; preds['Ridge']=rr.predict(Xte)

# 2. Random Forest
rf=RandomForestRegressor(n_estimators=200,max_depth=8,min_samples_leaf=5,
                          random_state=SEED,n_jobs=-1)
rf.fit(Xtr,ytr); models_trained['RandomForest']=rf; preds['RandomForest']=rf.predict(Xte)

# 3. XGBoost / GradientBoosting
if XGB:
    gb=xgb.XGBRegressor(n_estimators=300,max_depth=5,learning_rate=.05,
                          subsample=.8,colsample_bytree=.8,random_state=SEED,verbosity=0)
    gb_name='XGBoost'
else:
    gb=GradientBoostingRegressor(n_estimators=200,max_depth=4,learning_rate=.05,
                                  subsample=.8,random_state=SEED)
    gb_name='GradientBoosting'
gb.fit(Xtr,ytr); models_trained[gb_name]=gb; preds[gb_name]=gb.predict(Xte)

print(f"{'Model':<22}|{'RMSE':>8}|{'MAE':>8}|{'R²':>8}|{'MAPE%':>8}")
print("-"*60)
eval_res={}
for nm,pr in preds.items():
    rmse=np.sqrt(mean_squared_error(yte,pr))
    mae=mean_absolute_error(yte,pr)
    r2=r2_score(yte,pr)
    mape=np.mean(np.abs((yte-pr)/(yte+1e-6)))*100
    eval_res[nm]={'RMSE':rmse,'MAE':mae,'R2':r2,'MAPE':mape,'preds':pr}
    print(f"{nm:<22}|{rmse:>8.2f}|{mae:>8.2f}|{r2:>8.4f}|{mape:>7.2f}%")

best_mn=min(eval_res,key=lambda k:eval_res[k]['RMSE'])
print(f"\n🏆 Best model: {best_mn}")
````

# Section 10: Spatial Clustering for Delivery Zone Design
## 10.1 Theory

| Algorithm | Key Parameter | Strength | Weakness |
|-----------|--------------|---------|---------|
| K-Means | K (clusters) | Fast, convex zones | Requires K, circular zones |
| DBSCAN | ε, MinPts | No K needed, handles noise | Sensitive to parameters |
| Hierarchical Ward | K (cut) | Dendrogram insight | $O(n^2)$ memory |

**Silhouette score** $s(i)=\frac{b(i)-a(i)}{\max(a(i),b(i))}$ ranges −1 to +1; higher is better.  
**Davies-Bouldin index**, lower is better: $DB=\frac{1}{n}\sum_{i}\max_{j\neq i}\frac{s_i+s_j}{d_{ij}}$


````python
# Section 10 | Spatial Clustering
from sklearn.metrics import silhouette_score

coords=customers[['latitude','longitude']].values
cs=StandardScaler().fit_transform(coords)

# Optimal K
k_range=range(2,12)
inertias=[]; sils=[]
for k in k_range:
    km=KMeans(n_clusters=k,random_state=SEED,n_init=10)
    lbs=km.fit_predict(cs)
    inertias.append(km.inertia_); sils.append(silhouette_score(cs,lbs))

K_OPT=list(k_range)[np.argmax(sils)]
print(f"Optimal K = {K_OPT} (max silhouette = {max(sils):.4f})")

# Apply methods
km_f=KMeans(n_clusters=K_OPT,random_state=SEED,n_init=10)
km_lbl=km_f.fit_predict(cs)

db=DBSCAN(eps=.25,min_samples=8)
db_lbl=db.fit_predict(cs)
n_db=len(set(db_lbl))-(1 if -1 in db_lbl else 0)
n_noise=(db_lbl==-1).sum()

hc=AgglomerativeClustering(n_clusters=K_OPT,linkage='ward')
hc_lbl=hc.fit_predict(cs)

print(f"K-Means sil={silhouette_score(cs,km_lbl):.4f}")
print(f"DBSCAN: {n_db} clusters, {n_noise} noise points")
print(f"Hierarchical sil={silhouette_score(cs,hc_lbl):.4f}")

customers['zone_km']=km_lbl; customers['zone_db']=db_lbl; customers['zone_hc']=hc_lbl
````

# Section 11: Supply Chain Resilience Analysis
## 11.1 Theoretical Framework

Supply chain resilience quantifies a system's ability to **anticipate, absorb, adapt, and recover** from disruptions.

**Resilience Triangle** (Bruneau et al., 2003):
$$R = \int_{t_0}^{t_1}[100-Q(t)]\,dt$$

where $Q(t)$ is system performance during disruption window $[t_0,t_1]$.

### Network Vulnerability Metrics

| Metric | Definition |
|--------|-----------|
| **Global Efficiency** | $E=\frac{1}{n(n-1)}\sum_{i\neq j}\frac{1}{d_{ij}}$ |
| **Connectivity** | Fraction of nodes in largest connected component |
| **Robustness** | Relative efficiency after node removal |

### Disruption Scenarios Simulated
1. **Targeted Attack**: remove highest-betweenness nodes first (deliberate sabotage)
2. **Random Failure**: remove random nodes (equipment failure, weather)
3. **Warehouse Closure**: close each warehouse and measure service degradation
4. **Flood Scenario**: edge weight ×5 in a geographic band


````python
# Section 11 | Resilience Analysis
def net_efficiency(G):
    if G.number_of_nodes()<2: return 0.
    nodes=list(G.nodes())[:min(50,G.number_of_nodes())]
    tot=0; cnt=0
    for src in nodes:
        try:
            L=nx.single_source_shortest_path_length(G,src)
            for t,d in L.items():
                if t!=src and d>0: tot+=1./d; cnt+=1
        except: pass
    return tot/cnt if cnt>0 else 0.

def connectivity(G):
    if G.number_of_nodes()==0: return 0.
    comps=list(nx.weakly_connected_components(G) if G.is_directed()
               else nx.connected_components(G))
    return max(len(c) for c in comps)/G.number_of_nodes()

base_eff=net_efficiency(G_sample); base_con=connectivity(G_sample)
print(f"Baseline — Efficiency: {base_eff:.4f} | Connectivity: {base_con:.4f}")

# Attack simulation
fracs=np.linspace(0,.5,12)
eff_t=[]; eff_r=[]; con_t=[]; con_r=[]
sorted_tgt=[n for n,_ in sorted(betweenness_centrality.items(),
                                  key=lambda x:x[1],reverse=True)
            if n in G_sample]

for fr in fracs:
    nr=int(fr*len(G_sample))
    Ga=G_sample.copy(); Ga.remove_nodes_from([n for n in sorted_tgt[:nr] if n in Ga])
    eff_t.append(net_efficiency(Ga)/(base_eff+1e-9)); con_t.append(connectivity(Ga))
    Gr=G_sample.copy()
    rn=np.random.choice(list(Gr.nodes()),size=min(nr,Gr.number_of_nodes()),replace=False)
    Gr.remove_nodes_from(rn)
    eff_r.append(net_efficiency(Gr)/(base_eff+1e-9)); con_r.append(connectivity(Gr))

print("Attack simulation done")

# Warehouse closure impact
cl_res=[]
for ci in range(len(warehouses)):
    op=[i for i in range(len(warehouses)) if i!=ci]
    md2=DM[:,op].min(axis=1); wdist=(md2*demands).sum()
    cl_res.append({'wh':warehouses.iloc[ci]['name'],'wdist':wdist,
                   'maxd':md2.max(),'uncov':(md2>10).sum(),
                   'score':wdist/1000+(md2>10).sum()*10})
cl_df=pd.DataFrame(cl_res).sort_values('score',ascending=False)
print("Top 5 critical warehouses (closure impact):")
print(cl_df.head(5)[['wh','wdist','maxd','uncov','score']].round(2).to_string(index=False))
````

# Section 12: Multi-Criteria Decision Analysis (MCDA)
## 12.1 Analytic Hierarchy Process (AHP)

AHP (Saaty, 1980) structures complex decisions through pairwise comparisons.

**Consistency check:** $CR = CI/RI$ where $CI=(\lambda_{max}-n)/(n-1)$.  
$CR<0.10$ indicates acceptable consistency.

### Criteria and AHP Weights

| Criterion | Weight | Rationale |
|-----------|--------|-----------|
| Road Accessibility | ~35% | Primary cost driver |
| Demand Proximity | ~28% | Service level |
| Land Cost | ~18% | Capital expenditure |
| Flood Risk | ~12% | Resilience |
| Labour Availability | ~7% | Operational continuity |


````python
# Section 12 | AHP Analysis
crit_names=['Road Access','Demand Prox','Land Cost','Flood Risk','Labour']
AHP=np.array([[1,2,3,4,5],[1/2,1,2,3,4],[1/3,1/2,1,2,3],
              [1/4,1/3,1/2,1,2],[1/5,1/4,1/3,1/2,1]])

def ahp_w(M):
    n=M.shape[0]
    nrm=M/M.sum(0); w=nrm.mean(1)
    lmax=(M@w/w).mean(); CI=(lmax-n)/(n-1)
    RI={3:.58,4:.90,5:1.12,6:1.24,7:1.32,8:1.41,9:1.45,10:1.49}
    CR=CI/RI.get(n,1.49)
    return w,CR

wts,CR=ahp_w(AHP)
print("AHP Weights:"); 
for c,w in zip(crit_names,wts): print(f"  {c:<18}: {w:.4f} ({w*100:.1f}%)")
print(f"Consistency Ratio CR={CR:.4f} {'OK' if CR<.1 else 'Revise'}")
````

````python
# Section 12 | Suitability Scoring
def suitability(wdf,cdf,ws,dm):
    n=len(wdf); sc=pd.DataFrame({'id':wdf['warehouse_id']})
    sc['road']=wdf['road_access_score'].values
    sc['demand']=1-MinMaxScaler().fit_transform(dm.mean(0).reshape(-1,1)).ravel()
    sc['cost']=1-MinMaxScaler().fit_transform(wdf['fixed_cost_annual_usd'].values.reshape(-1,1)).ravel()
    np.random.seed(42)
    sc['flood']=1-np.random.beta(2,5,n)
    cc=np.array([haversine(np.array([r.latitude]),np.array([r.longitude]),
                            np.array([-1.2921]),np.array([36.8219]))[0,0]
                 for r in wdf.itertuples()])
    sc['labour']=1-MinMaxScaler().fit_transform(cc.reshape(-1,1)).ravel()
    cols=['road','demand','cost','flood','labour']
    for c in cols: sc[c]=MinMaxScaler().fit_transform(sc[[c]]).ravel()
    sc['score']=sum(ws[i]*sc[cols[i]] for i in range(5))
    sc['rank']=sc['score'].rank(ascending=False).astype(int)
    return sc.sort_values('rank').merge(wdf[['warehouse_id','name','latitude','longitude']],
                                        left_on='id',right_on='warehouse_id')

suit=suitability(warehouses,customers,wts,DM)
print("Top 5 warehouse sites:")
print(suit[['rank','name','road','demand','cost','flood','labour','score']].head(5).round(4).to_string(index=False))
````

Ensemble methods (Random Forest, XGBoost) substantially outperform Ridge Regression, capturing non-linear weekly and annual demand cycles. Lag features at t-7 and t-28 are the dominant predictors, confirming strong demand periodicity.

### Finding 4: Spatial Demand Structure
ESDA confirms significant positive Moran's I, validating demand clustering. DBSCAN identifies organic delivery zones without requiring K specification, outperforming K-Means for non-convex patterns.
